```python
import requests
```

```python
def request_end_point(url_request, request_method, input_data_default = {'title': 'recommendation', 'body':'motorcycle', 'userId':12} ):
    if request_method == 'GET' :
        # Penggunaan Request untuk mengirimkan permintaan (request) - GET
        html_dom_get = requests.get(url_request)
        return html_dom_get.json()
    else:
        # Penggunaan Request untuk mengirimkan permintaan (request) - POST
        input_data = input_data_default
        html_dom_post = requests.post(url_request, data = input_data)
        return html_dom_post.json()
```

```python
def check_end_point_correct_data_type(request_json, set_data_type):
    # Jika tipe dict berarti data hanya satu, namun jika tidak data lebih dari satu
    if type(request_json) == dict:
        request_json = [request_json]
        
    count_request = 0
    request_in_data_not_same_type = []
    request_in_data_same_type = []
    for json_dict in request_json:
        if set_data_type <= set(json_dict):
            if isinstance(json_dict['userId'], int) and isinstance(json_dict['id'], int) and isinstance(json_dict['title'], str) and isinstance(json_dict['body'], str) :
                request_in_data_same_type.append(json_dict)
            else:
                request_in_data_not_same_type.append(json_dict)
            print('Key: userId - Value: ',json_dict['userId'],' == ',type(json_dict['userId']))
            print('Key: id - Value: ',json_dict['id'],' == ',type(json_dict['id']))
            print('Key: title - Value: ',json_dict['title'],' == ',type(json_dict['title']))
            print('Key: body - Value: ',json_dict['body'],' == ',type(json_dict['body']))
            print('Hasil Pengecheckan : ',isinstance(json_dict['userId'], int) and isinstance(json_dict['id'], int) and isinstance(json_dict['title'], str) and isinstance(json_dict['body'], str))
            print('\n')
            count_request += 1
        
    if count_request < 0:
        return {'MESSAGE':'Tidak ada end point sesuai dengan type data', 'JUMLAH JSON':count_request, 'HASIL JSON SESUAI TYPE DATA': request_in_data_same_type, 'HASIL JSON TIDAK SESUAI TYPE DATA': request_in_data_not_same_type} 
    else:
        return {'MESSAGE':'Ada end point sesuai dengan type data', 'JUMLAH JSON':count_request, 'HASIL JSON SESUAI TYPE DATA': request_in_data_same_type, 'HASIL JSON TIDAK SESUAI TYPE DATA': request_in_data_not_same_type}
```

```python
if __name__ == "__main__" :
    URL_BASE = 'https://jsonplaceholder.cypress.io/posts'
    METHOD_REQ = 'GET' # atau POST
    
    #Input Data, jika dimasukkan ke methode 'request_end_point' maka input data akan sesuai masukkan. 
    # ... Jika tidak maka akan sesuai dengan default
#     input_data = {'title': 'sunt aut facere repellat provident occaecati excepturi optio reprehenderit', 
#                   'body':'quia et suscipit\nsuscipit recusandae consequuntur expedita et cum\nreprehenderit molestiae ut ut quas totam\nnostrum rerum est autem sunt rem eveniet architecto', 
#                   'userId':1}
#     json_data = request_end_point(URL_BASE, METHOD_REQ, input_data)
    json_data = request_end_point(URL_BASE, METHOD_REQ)
    
    set_data_type = {'title','body','userId','id'}
    check_end_point = check_end_point_correct_data_type(json_data, set_data_type)
    
    # KELUARAN (OUTPUT)
    print('End Point yang di check : ',set_data_type)
    print('Message :',check_end_point['MESSAGE'], ' - Hasil :',check_end_point['JUMLAH JSON'])
    print('Hasil Total Sesuai:', len(check_end_point['HASIL JSON SESUAI TYPE DATA']) )
    print('Hasil Total Tidak Sesuai :', len(check_end_point['HASIL JSON TIDAK SESUAI TYPE DATA']) )
#     #ATAU 
#     print(check_end_point)
```
